```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
```

```python
df=pd.read_csv('telcoo.csv')
```

```python
df.head()
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
df.head()
```

```python
x=df.drop('Churn',axis=1)
x
```

```python
y=df['Churn']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#decision tree classifier
```

```python
model_dt=DecisionTreeClassifier(criterion = "gini" , random_state=100,max_depth=6,min_samples_leaf=8)
```

```python
model_dt.fit(x_train,y_train)
```

```python
y_pred=model_dt.predict(x_test)
y_pred
```

```python
model_dt.score(x_test,y_test)
```

```python
print(classification_report(y_test,y_pred,labels=[0,1]))
```

```python
#the accuracy is bad because the data is imbalanced ,hence we use smoteenn
```

```python
sm = SMOTEENN()
x_resample,y_resample = sm.fit_resample(x,y)
```

```python
xr_train,xr_test,yr_train,yr_test=train_test_split(x_resample,y_resample,test_size=0.2)
```

```python
model_dt_smote=DecisionTreeClassifier(criterion='gini',max_depth=6,random_state=100,min_samples_leaf=8)
```

```python
model_dt_smote.fit(xr_train,yr_train)
```

```python
yr_pred=model_dt_smote.predict(xr_test)
yr_pred
```

```python
model_dt_smote.score(xr_test,yr_test)
```

```python
print(classification_report(yr_test,yr_pred,labels=[0,1]))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf_smote=RandomForestClassifier(n_estimators=100,criterion='gini',random_state=100,max_depth=6,min_samples_leaf=8)
```

```python
model_rf_smote.fit(xr_train,yr_train)
```

```python
yr_pred=model_rf_smote.predict(xr_test)
```

```python
model_rf_smote.score(x_test,y_test)
```

```python
print(classification_report(yr_test,yr_pred,labels=[0,1]))
```

```python
print(confusion_matrix(yr_test,yr_pred))
```
